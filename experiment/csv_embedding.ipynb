```python
import re
import pandas as pd

df = pd.read_csv('문정고_menu.csv')
menu_info = df['Menu'].replace({'\n':'','\t':''},regex=True)
menu_info = menu_info.replace(['.',''],pd.NA)
menu_info = menu_info.dropna()
menu_info = menu_info.reset_index(drop = True)
menus = []

pattern = re.compile(r"<.*?>|\★.*?\★")
for i in menu_info:
    menu = re.sub(pattern, '', i)
    menu = menu.replace(' ',',').strip()
    menu = menu.split(',')
    
    menus += menu

print(menus)
print(len(menus)) # 줄이기 전 4655
menus = list(set(menus))
print(menus)
print(len(menus)) # 중복 제거 후 1183


```

## 전체 csv 파일에서 적용

```python
import pandas as pd
df1 = pd.read_csv('문정고_menu.csv')
df2 = pd.read_csv('선유고_menu.csv')
df3 = pd.read_csv('신현고_menu.csv')
df4 = pd.read_csv('월계고_menu.csv')
df5 = pd.read_csv('자운고_menu.csv')

```

```python
print(len(df1))
print(len(df3))
print(len(df5))
```

```python


df = pd.concat([df1, df3 ,df5], axis =0)
df = df.reset_index(drop=True)
print(df.head())
print(len(df))
print(df['Menu'].head())
```

```python

def menu_clean(df):
    menu_info = df['Menu'].replace({'\n':'','\t':''},regex=True)
    menu_info = menu_info.replace(['.',''],pd.NA)
    menu_info = menu_info.dropna()
    menu_info = menu_info.reset_index(drop = True)
    menus = []

    pattern = re.compile(r'<.*?>|\★.*?\★|\([^)]*\)')
    pattern = re.compile(r'\(\d+(\.\d+)*\)|\d+(\.\d+)*|\*|-|<.*?>|\★.*?\★')
    for i in menu_info:
        menu = re.sub(pattern, '', i)
        menu = menu.replace('.','').strip()
        menu = menu.replace(' ',',').strip()
        menu = menu.split(',')
        
        menus += menu
    menus = [menu for menu in menus if menu != '' and menu != '메뉴없음' and menu != '///' and menu != '.']

    return menus

menus = menu_clean(df)
print(menus)
print(len(menus)) # 중복 제거 전 12382 (3학교)
print(len(set(menus))) # 중복 제거 후 3580(3학교-문정,신현,자운)
```

```python
# 3학교 메뉴 형태소 분석

from collections import Counter
kkma = Kkma()
keywords = []
for menu_name in menus:
    if menu_name:
        nouns = kkma.nouns(menu_name)
        keywords.extend([m for m in nouns if len(m) > 1])

keyword_counter = Counter(keywords)
most_common_keywords = keyword_counter.most_common()
most_keywords = []

for keyword, count in most_common_keywords:
    most_keywords.append(keyword)
    print(f'{keyword} : {count}')

print(len(most_common_keywords))
```

most_common_keywords - 

1) 3학교 통틀어서 메뉴 추출 후 중복 제거

2) 형태소별로 분석 (이때는 중복 제거 x)

3) (keyword - 등장 횟수) 형태로 리스트에 담기 



1 ) 에서 나온 메뉴를 3)에서 추출한 리스트의 0번째 원소로 replace

```python
def clean_file(df_ex):
    new_df = df_ex[['File Name','Menu']].copy()
    new_df.loc[:,'Menu'] = new_df['Menu'].replace({'\n':'','\t':'','메뉴없음':''},regex=True)
    new_df.loc[:,'Menu'] = new_df['Menu'].replace(['.',''],pd.NA)
    new_df = new_df.dropna(subset=['Menu'])
    new_df.loc[:,'Menu'] = new_df['Menu'].replace(' ',',', regex = True)
    new_df = new_df.reset_index(drop=True)
    return new_df

def menu_change(df_ex):
    df_ex2 = df_ex.copy()
    for i,row in df_ex2.iterrows():
        new_menu = []
        menus = row['Menu'].split(',')
        for word in menus:
            for keyword in most_keywords[20:]:
                if keyword in word:
                    new_menu.append(keyword)
                    break
            else:
                new_menu.append(word)
        df_ex2.at[i,'Menu']=','.join(new_menu)

    return df_ex2
```

```python
df1 = pd.read_csv('문정고_menu.csv')
ori_df = clean_file(df1)
new_df = menu_change(ori_df)

print(ori_df)
print(new_df)
```

```python
df5 = pd.read_csv('자운고_menu.csv')
ori_df = clean_file(df5)
new_df = menu_change(ori_df)

print(ori_df.head(10))
print(new_df.head(10))
```

```python
new_df.to_csv('자운고_menu_수정.csv', index = False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df = pd.read_csv('문정고_menu.csv')
menu_info = df['Menu'].replace({'\n':'','\t':''},regex=True)
ex = str(menu_info.replace(',',' '))
text = ex.split()

menu_names = text

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(menu_names)
word_index = tfidf_vectorizer.vocabulary_

for i, menu_name in enumerate(menu_names):
    tfidf_scores = tfidf_matrix[i,:].toarray().flatten()
    max_tfidf_index = tfidf_scores.argmax()
    keyword = list(word_index.keys())[list(word_index.values()).index(max_tfidf_index)]
    print(f"{menu_name} -> {keyword}")
```
